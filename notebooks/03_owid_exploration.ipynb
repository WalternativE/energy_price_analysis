```python
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = (10, 8)

import seaborn as sns
sns.set_theme(style="whitegrid")
```

```python
owid_data = pd.read_csv("../data/owid/owid-energy-data.csv")
owid_data
```

```python
countries_in_europe = pd.read_csv("../data/misc/countries_in_europe.csv").drop(columns=["zoom", "abbreviation", "continent"])
```

```python
owid_europe = owid_data[owid_data["iso_code"].isin(countries_in_europe["ISO alpha 3"])]
owid_europe = owid_europe[owid_europe["year"] >= 2000]
owid_europe
```

```python
owid_europe_members = set(owid_europe["iso_code"].unique())
europe_members = set(countries_in_europe["ISO alpha 3"].unique())
europe_members.difference(owid_europe_members)
```

```python
sns.lineplot(data=owid_europe.sort_values(by="wind_share_energy", ascending=False), x="year", y="wind_share_energy")
plt.show()
```

```python
sns.barplot(
    data=(
        owid_europe.drop(columns=["year"])
            .groupby(["country"])
            .mean()
            .reset_index()
            .sort_values(by="wind_share_energy", ascending=False)
            .head(n=20)
    ),
    x="country",
    y="wind_share_energy",
)
plt.xticks(rotation=45)
plt.show()
```

```python
# there is no data for 2020 yet
top_highest_windshare_2019 = (
    owid_europe[owid_europe["year"] == 2019]
        .sort_values(by="wind_share_energy", ascending=False)
        .head(n=20)["country"]
        .to_list()
)

sns.lineplot(
    data=owid_europe[owid_europe["country"].isin(top_highest_windshare_2019)],
    x="year",
    y="wind_share_energy",
    style="country",
    hue="country",
)
```

```python
sns.barplot(
    data=(
        owid_europe[owid_europe["year"] == 2019]
            .sort_values(by="wind_share_energy", ascending=False)
            .head(n=20)
    ),
    x="country",
    y="wind_share_energy",
)
plt.xticks(rotation=45)
plt.show()
```
