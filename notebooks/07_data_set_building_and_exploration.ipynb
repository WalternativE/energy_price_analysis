```python
import pandas as pd
import numpy as np

import holoviews as hv
from holoviews import opts
import hvplot.pandas
import matplotlib.pyplot as plt

%matplotlib inline
plt.rcParams["figure.figsize"] = (10, 8)

import seaborn as sns
sns.set_theme(style="whitegrid")
```

```python
entsoe_capacities = pd.read_csv("../data/entsoe/capacity_2015_2021.csv").drop("Unnamed: 0", axis=1)
capacities_of_interest = [col for col in entsoe_capacities if ("gas" in col.lower() or "wind" in col.lower())]
entsoe_cols = capacities_of_interest + ["year", "country"]
entsoe_capacities = entsoe_capacities[entsoe_cols]
entsoe_capacities.head(5)
```

```python
entsoe_capacities["total_capacity"] = entsoe_capacities[capacities_of_interest].sum(axis=1)
for capacity in capacities_of_interest:
    entsoe_capacities[f"Relative {capacity}"] = entsoe_capacities[capacity] / entsoe_capacities["total_capacity"]
entsoe_capacities.head(5)
```

```python
plots = [
    entsoe_capacities[entsoe_capacities["country"] == country].hvplot(
        x="year", y=[f"Relative {capacity}" for capacity in capacities_of_interest], title=country
    ) for country in entsoe_capacities["country"].unique()
]
hv.Layout(plots).cols(2).opts(norm={"axiswise": True})
```
